# MFCC

```python
# Importamos las librerías que hemos estado usando
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np

# Importamos el audio con el que hemos estado trabjando

# Vamos a cargar nuestro archivo de audio de prueba 
audio_path = r"C:\Users\Diego Max.000\Desktop\SS\AUDIOS\harvard.wav"

# Vamos a escuchar el audio
ipd.Audio(audio_path)

# Cargamos el audio con librosa
y, sr = librosa.load(audio_path) 
```

```python
# numero de muestras
y.shape
```

```python
# Extraer los MFCC es muy facil con librosa pues ya hay una función 
mfcc_y = librosa.feature.mfcc(y = y, n_mfcc=13, sr =sr)

mfcc_y.shape
```

```python
# Ahora vamos a ver los mfcc
plt.figure(figsize=(25,10))
librosa.display.specshow(mfcc_y,
                         x_axis="time",
                         sr = sr)
plt.colorbar(format = "*%.2f")
plt.show()
```

```python
# Ahora vamos a calcular los delta y delta-delta de los MFCC

delta_mfcc_y = librosa.feature.delta(mfcc_y)
delta2_mfcc_y = librosa.feature.delta(mfcc_y, order = 2)
```

```python
# Vamos a ver los deltas
# Ahora vamos a ver los mfcc
plt.figure(figsize=(25,10))
librosa.display.specshow(delta_mfcc_y,
                         x_axis="time",
                         sr = sr)
plt.colorbar(format = "*%.2f")
plt.show()
```

```python
# Ahora vamos a ver los mfcc
plt.figure(figsize=(25,10))
librosa.display.specshow(delta2_mfcc_y,
                         x_axis="time",
                         sr = sr)
plt.colorbar(format = "*%.2f")
plt.show()
```

```python
# Ahora vamos a unir los coeficientes con los deltas para obtener
# un solo vector que contenga la información más relvante

concat_mfcc = np.concatenate((mfcc_y,delta_mfcc_y, delta2_mfcc_y))
concat_mfcc.shape
```
