```python
import librosa

import matplotlib.pyplot as plt
import numpy as np
```

```python
def set_default(figsize=(10,10), dpi=100):
    plt.style.use([ 'dark_background', 'bmh'])
    plt.rc('axes', facecolor='k')
    plt.rc('figure', facecolor='k')
    plt.rc('figure', figsize=figsize, dpi=dpi)
```

```python
set_default(figsize=(16, 8))
```

```python
audio = r"C:\Users\Diego Max.000\Desktop\SS\AUDIOS\harvard.wav"
y, sampling_rate = librosa.load(audio)
sampling_rate
```

```python
from IPython.display import display, Audio
Audio(y, rate=sampling_rate)
```

```python
T = y.size / sampling_rate
dt = 1 / sampling_rate
t = np.r_[0:T:dt]
```

```python
print(
    f'y[t] tiene {y.size} muestras',
    f'la frecuencia de muestreo es {sampling_rate * 1e-3}kHz',
    f'y(t) tiene {T:.1f}s '
    , sep='\n')
```

```python
plt.plot(t, y)
plt.figure()
plt.plot(t, y)
plt.xlabel('tiempo [s]')
plt.ylabel('amplitud[/]')
plt.title(r'$y(t)$', size=20);
```

```python
X = librosa.stft(y)
X_dB = librosa.amplitude_to_db(np.abs(X))
X = librosa.stft(y)
```

```python
from librosa.display import specshow
specshow(X_dB, sr=sampling_rate, x_axis='time', y_axis='hz')
plt.xlabel('tiempo [s]')
plt.ylabel('frecuencia [Hz]')
plt.ylim(top=12000)
plt.colorbar()
plt.grid(True)
```
