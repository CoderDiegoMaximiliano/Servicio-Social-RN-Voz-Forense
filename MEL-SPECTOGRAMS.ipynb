# MEL SPECTOGRAMS

```python
# Importamos las librerías que hemos estado usando
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np

# Importamos el audio con el que hemos estado trabjando

# Vamos a cargar nuestro archivo de audio de prueba 
audio_path = r"C:\Users\Diego Max.000\Desktop\SS\AUDIOS\harvard.wav"

# Vamos a escuchar el audio
ipd.Audio(audio_path)

# Cargamos el audio con librosa
y, sr = librosa.load(audio_path) 
```

## MEL FILTER BANKS

```python
# Extremos los bancos de filtros de mel con librosa

filter_banks = librosa.filters.mel(n_fft=2048, sr = 22050, n_mels=10)
```

```python
filter_banks.shape

# dim = (#mel_banks, Nyquist_Frq +1)
# Nyquist_Freq = (nfft /2) + 1

```

```python
plt.figure(figsize=(26,10))
librosa.display.specshow(filter_banks,
                         sr = sr,
                         x_axis="linear")
plt.colorbar(format="*%+2.f")
plt.show
```

```python
mel_spectogram = librosa.feature.melspectrogram(y = y, 
                                                sr=sr, 
                                                n_fft=2048, 
                                                hop_length=512, 
                                                n_mels=10)
```

```python
mel_spectogram.shape

# (n_mel_banks, # frames)
```

```python
# Ahora vamos a hacer un log-mel espectograma

log_mel_spectogram = librosa.power_to_db(mel_spectogram)

log_mel_spectogram.shape
```

```python
# Ahora vamos a verlo
plt.figure(figsize=(25,10))
librosa.display.specshow(log_mel_spectogram,
                         x_axis="time",
                         y_axis="mel",
                         sr = sr)
plt.colorbar(format="*%.2f")
plt.show
```
